```python
import numpy as np 

# Criando um vetor de dados aleatórios
data = np.random.rand(1000,2) # matriz de 1000 linhas e 2 colunas

# Verificando o tamanho do vetor
print(data.shape)
print(np.size(data)) # quantidade de elementos 

# Print de uma amostra dos dados
print(data[0:10]) # retorna os primeiros dez elementos da sequência 
print(data[0:10].shape) 
```

```python
# Usando Monte-Carlo 
dataMax = 1000
dataMin = 10
data = (dataMax - dataMin)*data + dataMin  
print(data[0:10]) 
# A simulação de Monte-Carlo é uma ferramenta para prever cenários 
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3) # três gráficos na mesma figura
ax[0].plot(data) # todos os elementos de data
ax[1].plot(data[:,0]) # primeira coluna de data
ax[2].plot(data[:,1]) # segunda coluna de data 
print(type(data))
```

```python
# Criando uma função
def noiseCreation(path, year):
    """Esta função cria um ruido com sazonalidade.
    Inputs:
        path = asdasdas
        year = asdasdasdas

    Criado por Leonardo
    """
    
    # Criando um dataframe
    df = pd.DataFrame()
    
    # Criando vertor temporal
    dtime = np.arange(np.datetime64(str(year)+"-01-01"), np.datetime64(str(year+1)+"-01-01"), np.timedelta64(1, "h"))
    
    # Monstrando um dataframe
    df['datetime'] = dtime
        
    # Datetime como index do dataframe
    df = df.set_index(df['datetime'])
    
    # Criando colunas com ano, mês e dia
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    
    # Crio uma coluna random
    df['random'] = np.random.rand(df.shape[0])
    
    # Adicionar sazonalidade
    tt = np.arange(0,df.shape[0],1)
    print(tt.shape)
    df['sazonal'] = (np.cos(tt/(360*4)))
    df['timeSeries'] = 10*df['sazonal']+df['random']
    #df['sazonal'].plot()
    df['timeSeries'].plot(color='red')
    df.to_csv(path+'/noise_'+str(year)+'.csv')
    
    return df

# Foi criada uma função que faz o mesmo processo detalhado anteriormente
# Criar funções é melhor porque economiza linhas de código, basta chamar a função quando necessário, ao invés de detalhar todo o procedimento novamente
```
